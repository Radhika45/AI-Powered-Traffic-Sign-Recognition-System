```python
# 📌 Step 1: Import Required Libraries
import cv2
import numpy as np
import tensorflow as tf
import json
import os
from tensorflow.keras.preprocessing import image

# 📌 Step 2: Load the Trained Model
model_path = "../Real_Time_Inference/model.keras"

if os.path.exists(model_path):
    model = tf.keras.models.load_model(model_path)
    print("✅ Model loaded successfully!")
    print(f"📌 Model expects input shape: {model.input_shape}")  # Debugging line
else:
    print("❌ ERROR: model.keras NOT found! Check the file path.")
    exit()

# 📌 Step 3: Load Class Labels
labels_path = "../Real_Time_Inference/labels.json"

if os.path.exists(labels_path):
    with open(labels_path, "r") as f:
        try:
            labels = json.load(f)
            print("✅ Labels loaded successfully!")
        except json.JSONDecodeError:
            print("❌ ERROR: labels.json is empty or not properly formatted.")
            exit()
else:
    print("❌ ERROR: labels.json NOT found! Check the file path.")
    exit()

# 📌 Step 4: Initialize Webcam
cap = cv2.VideoCapture(0)  # 0 for default webcam

if not cap.isOpened():
    print("❌ ERROR: Could not access webcam!")
    exit()

print("📷 Webcam started. Press 'q' to exit.")

# 📌 Step 5: Define Input Size (Model's Expected Size)
IMG_SIZE = (48, 48)  # Update to match your model's expected input size

while True:
    ret, frame = cap.read()
    if not ret:
        print("❌ ERROR: Could not capture frame.")
        break

    # 📌 Step 6: Preprocess the Frame
    img = cv2.resize(frame, IMG_SIZE)  # Resize to (48, 48)
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize

    # 📌 Step 7: Make Prediction
    predictions = model.predict(img)
    class_index = np.argmax(predictions)
    confidence = np.max(predictions)

    # 📌 Step 8: Get Predicted Label
    sign_label = labels.get(str(class_index), "Unknown Sign")

    # 📌 Step 9: Display Result on Frame
    text = f"{sign_label} ({confidence:.2f})"
    cv2.putText(frame, text, (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # 📌 Step 10: Show the Output
    cv2.imshow("Traffic Sign Recognition", frame)

    # Exit when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# 📌 Step 11: Release Resources
cap.release()
cv2.destroyAllWindows()
print("🛑 Real-time recognition stopped.")
```
